```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('/content/Ensayo de Compresion botella de gaseosa_1.xlsx')
```

```python
df
```

```python
df.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
df = df.rename(columns={'Unnamed: 1': 'Tiempo (s)', 'Unnamed: 2': 'Desplazamiento (mm)', 'Unnamed: 3': 'Fuerza (kN)'})
```

```python
df
```

```python
filas_a_eliminar=[0,1,2,3,4,5,6,7]

df.drop(filas_a_eliminar)
```

```python
df=df.drop(filas_a_eliminar)
```

```python
df
```

```python
df['Desplazamiento (mm)'] = df['Desplazamiento (mm)'] / 1000
df['Fuerza (kN)'] = df['Fuerza (kN)'] * 1000
```

```python
df = df.rename(columns={'Desplazamiento (mm)': 'Desplazamiento (m)', 'Fuerza (kN)': 'Fuerza (N)'})
df
```

```python
df.isna().sum()
```

```python
df.duplicated()
```

```python
df.info()
```

```python
df['Tiempo (s)'] = pd.to_numeric(df['Tiempo (s)'])
df['Desplazamiento (m)'] = pd.to_numeric(df['Desplazamiento (m)'])
df['Fuerza (N)'] = pd.to_numeric(df['Fuerza (N)'])
```

```python
df.info()
```

```python
plt.figure(figsize=(15,7))
sns.lineplot(x=df["Desplazamiento (m)"],y=df['Fuerza (N)'])
plt.xlabel("Desplazamiento (m)")
plt.ylabel("Fuerza (N)")
plt.title("Grafica de Fuerza vs Desplazamiento")
```

```python
from scipy.integrate import simps
```

```python
x = df['Desplazamiento (m)']
y = df['Fuerza (N)']

area_bajo_curva = simps(y, x)
print(f"Área bajo la curva: {area_bajo_curva:10f}")
```
