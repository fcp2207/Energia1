```python
from scipy.integrate import simps
```

```python
x = botella_400_ml['Desplazamiento (m)']
y = botella_400_ml['Fuerza (N)']

area_bajo_curva = simps(y, x)
print(f"Energía necesaria para compactar botella de 400 ml (J): {area_bajo_curva:10f}")

x_1 = gatorade_1['Desplazamiento (m)']
y_1 = gatorade_1['Fuerza (N)']

area_bajo_curva_1 = simps(y_1, x_1)

x_2 = gatorade_2['Desplazamiento (m)']
y_2 = gatorade_2['Fuerza (N)']

area_bajo_curva_2 = simps(y_2, x_2)

Suma_Energia_Gatorade = area_bajo_curva_1 + area_bajo_curva_2
print(f"Energía necesaria para compactar botella de Gatorade (J): {Suma_Energia_Gatorade:10f}")

x_3 = coca_cola_botella['Desplazamiento (m)']
y_3 = coca_cola_botella['Fuerza (N)']

area_bajo_curva_3 = simps(y_3, x_3)
print(f"Energía necesaria para compactar botella de Coca-Cola de 400 ml (J): {area_bajo_curva_3:10f}")

x_4 = coca_cola_lata['Desplazamiento (m)']
y_4 = coca_cola_lata['Fuerza (N)']

area_bajo_curva_4 = simps(y_4, x_4)
print(f"Energía necesaria para compactar lata de Coca-Cola de 225 ml (J): {area_bajo_curva_4:10f}")


Suma_1 = area_bajo_curva + Suma_Energia_Gatorade + area_bajo_curva_3 + area_bajo_curva_4
Promedio = Suma_1 / 4
resultado_formateado= "{:.2f}".format(Promedio)
print("Energia promedio de compactación (J):",resultado_formateado)

Energía = resultado_formateado
Energía = float(resultado_formateado)
Distancia = 0.14
A = Energía/Distancia
Peso = A/9.81
print("El peso necesario para compactar la botella es :",Peso,"(Kg)")
```
